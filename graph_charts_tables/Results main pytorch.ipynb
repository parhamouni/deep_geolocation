```python
import pandas as pd
import os
from glob import glob
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
filenames_list = glob('./results/*.txt')
```

```python
results_list=[]
for address in filenames_list:
    results_list.append(pd.read_csv(address))
```

```python
df_results = pd.concat(results_list,ignore_index=True)
```

```python
df_results = df_results.drop(columns=['Unnamed: 0'])
```

```python
for ind, row in df_results.iterrows():
    if type(row.fraction)==str:
        df_results.loc[ind,'fraction'] = float(row.fraction[1:-1])
df_results.sort_values(['fraction','p','q'],inplace=True)
```

```python
df_results.reset_index(inplace=True,drop = True)
```

```python
df_results
```

```python
## seeing best results by model
g = df_results.groupby(['model_name'])
```

```python
fig, ax = plt.subplots()
for key, grp in g:
    ax = grp.plot(ax=ax, x='fraction', y='median', label=key, figsize = (15,15))

plt.legend(loc='best')
plt.show()
```

From this diagram we understand that neural nets and Linear SVM are the probabably the best ones

Now lets find out the doc2vec effect on neural networks

```python
df_nn = df_results[df_results.model_name=='Neural Net']
```

```python
g = df_nn.groupby('d2v')
```

```python
fig, ax = plt.subplots()
for key, grp in g:
    ax = grp.plot(ax=ax, x='fraction', y='median', label=key)

plt.legend(loc='best')
plt.show()
```

As far as can be seen, the dataset with tfidf shows better results. 

Now lets see which p and q combination is the best

```python
df_nn_tfidf = df_nn[df_nn.d2v==0]
```

```python
g = df_nn_tfidf.groupby(['p','q'])
```

```python
fig, ax = plt.subplots()
for key, grp in g:
    ax = grp.plot(ax=ax, x='fraction', y='median', label=key, figsize = (15,15))

plt.legend(loc='best')
plt.show()
```
